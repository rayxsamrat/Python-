Q.10   What is the difference between implicit
      and explicit type casting ?
        - Type casting is nothing but converting one data type to another data type.It is of two type
            i. Implicit Type Casting :-It is also known as automatic type conversion.It is performed by the Python interpreter automatically,without any user intervention.
        Ex - x = 10  # integer
             y = 3.14  # float

       result = x + y  # implicit type casting: x is converted to float before addition

      print(result)  # Output: 13.14 (float)

            ii. Explicit Type Cating:-It is also known as type coercion,Pperformed by the programmer using built in type conversion functions like int(),float(),str(),etc
            Ex- x = 10.5  # float
       y = int(x)  # explicit type casting: x is converted to integer
       print(y)  # Output: 10 (integer)
